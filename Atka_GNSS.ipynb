```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd
import numpy as np
from bokeh.plotting import figure
from bokeh.io import output_notebook, show
from bokeh.models import Slope

output_notebook()
```

```python
atkaGnssPath = "C:\\Users\\jackg\\Downloads\\AB01_Atka_GNSS_Data.csv"
adakGnssPath = "C:\\Users\\jackg\\Downloads\\AB21_Adak_GNSS_Data.csv"
```

```python
atkaGNSS = pd.read_csv(atkaGnssPath, header = 0, skiprows = 36)
adakGNSS = pd.read_csv(adakGnssPath, header = 0, skiprows = 36)
```

```python
# Calculate best fit line for both Atka and Adak
# https://www.statisticshowto.com/probability-and-statistics/regression-analysis/find-a-linear-regression-equation/#definition

# Atka
atkaGNSS["JJdU"] = atkaGNSS["JJJJJ.JJJJ"] * atkaGNSS["dU"]
atkaGNSS["JJsq"] = atkaGNSS["JJJJJ.JJJJ"] ** 2
atkaGNSS["dUsq"] = atkaGNSS["dU"] ** 2

# Slope
atkaSlope = (atkaGNSS.shape[0]*atkaGNSS["JJdU"].sum() - atkaGNSS["JJJJJ.JJJJ"].sum()*atkaGNSS["dU"].sum()) /(atkaGNSS.shape[0] * atkaGNSS["JJsq"].sum() - atkaGNSS["JJJJJ.JJJJ"].sum()**2)

# Intercept
atkaInter = (atkaGNSS["dU"].sum()*atkaGNSS["JJsq"].sum() - atkaGNSS["JJJJJ.JJJJ"].sum()*atkaGNSS["JJdU"].sum()) / (atkaGNSS.shape[0]*atkaGNSS["JJsq"].sum() - atkaGNSS["JJJJJ.JJJJ"].sum()**2)

# Adak
adakGNSS["JJdU"] = adakGNSS["JJJJJ.JJJJ"] * adakGNSS["dU"]
adakGNSS["JJsq"] = adakGNSS["JJJJJ.JJJJ"] ** 2
adakGNSS["dUsq"] = adakGNSS["dU"] ** 2

# Slope
adakSlope = (adakGNSS.shape[0]*adakGNSS["JJdU"].sum() - adakGNSS["JJJJJ.JJJJ"].sum()*adakGNSS["dU"].sum()) /(adakGNSS.shape[0] * adakGNSS["JJsq"].sum() - adakGNSS["JJJJJ.JJJJ"].sum()**2)

# Intercept
adakInter = (adakGNSS["dU"].sum()*adakGNSS["JJsq"].sum() - adakGNSS["JJJJJ.JJJJ"].sum()*adakGNSS["JJdU"].sum()) / (adakGNSS.shape[0]*adakGNSS["JJsq"].sum() - adakGNSS["JJJJJ.JJJJ"].sum()**2)
```

```python
# Create Atka Plot
atkaP = figure(width=950, height=600, title = "Atka GNSS Station: AB01 (UNAVCO)")
atkaP.scatter(atkaGNSS["JJJJJ.JJJJ"], atkaGNSS["dU"], color = "firebrick", size = 1)
atkaReg = Slope(gradient=atkaSlope, y_intercept=atkaInter, line_color="black")
atkaP.add_layout(atkaReg)

# Create Adak Plot
adakP = figure(width=950, height=600, title = "Adak GNSS Station: AB01 (UNAVCO)")
adakP.scatter(adakGNSS["JJJJJ.JJJJ"], adakGNSS["dU"], color = "darkblue", size = 1)
adakReg = Slope(gradient=adakSlope, y_intercept=adakInter, line_color="black")
adakP.add_layout(adakReg)
```

```python
show(atkaP)
show(adakP)
```
